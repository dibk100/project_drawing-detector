```python
from PIL import Image
import pytesseract

# 테스트 데이터 로드
image_path = "./data/test01_re.JPG"
image = Image.open(image_path).convert("RGB")

# OCR 수행
text = pytesseract.image_to_string(image, lang='eng+kor')  # 언어 지정 가능: 'kor' 등

print("인식된 텍스트:")
print(text)
```

```python
import easyocr

# reader 생성 (lang: 영어 + 한국어)
reader = easyocr.Reader(['en', 'ko'])

# 이미지 경로
image_path = "./data/test01_rere.JPG"

# OCR 수행
results = reader.readtext(image_path)

# 결과 출력
for bbox, text, prob in results:
    print(f"텍스트: {text} | 신뢰도: {prob:.2f}")
```
